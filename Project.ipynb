```python
park = [[126.8141241,37.5876169], [126.8979808,37.5397924], [126.9939698,37.5102684], [127.0165227,37.5272945], [127.081523,37.5179568]
         ,[127.1194287,37.5499881],[126.8721985,37.5649224],  [126.8986955,37.555875], [126.9736383,37.5159618], [127.0691564,37.5291008]]

def dist_park(x, y):
    z = []
    for i in range(10):
        k = ((114*(x - park[i][0]))**2 + ((92*(y - park[i][1]))**2))**(1.0/2.0)
        z.append(k)
    return min(z)

def dist_myeongdong(x, y):
    k = ((114*(x - 126.9653337))**2 +(92*(y - 37.5634056))**2)**(1.0/2.0) 
    return k
    
def dist_gangnam(x, y):
    k = ((114*(x - 127.0083444))**2 +(92*(y - 37.5002943))**2)**(1.0/2.0) 
    return k
```

```python
ds_pk = []
ds_my = []
ds_gn = []

for i in range(4493):
    x = float(df.ix[i,2][1:-2].split(', ')[0])
    y = float(df.ix[i,2][1:-2].split(', ')[1])
    ds_pk.append(dist_park(x, y))
    ds_my.append(dist_myeongdong(x, y))
    ds_gn.append(dist_gangnam(x, y))
```

```python
copy_df = df.copy()
copy_df
```

```python
data1 = [ds_pk[i] for i in range(4493)]
data2 = [ds_my[i] for i in range(4493)]
data3 = [ds_gn[i] for i in range(4493)]
temp1 = pd.Series(data1, name= 'distance_park')
temp2 = pd.Series(data2, name= 'distance_md')
temp3 = pd.Series(data3, name= 'distance_gn')
copy_df = copy_df.join(temp1)
copy_df = copy_df.join(temp2)
copy_df = copy_df.join(temp3)
```
