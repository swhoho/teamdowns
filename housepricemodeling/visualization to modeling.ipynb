* Variable Descriptions
    * address: Address of each house
    * animal: Whether or not you can raise animal in the house
    * building_floor: The total floors of the building where house is located 
    * d_to_bank: distance to closest bank
    * d_to_coffe: distance to closest cafe(m)
    * d_to_conv: distance(m)      
    * d_to_metro: distance to closest metro(m),
    * elevator : existence of elevator in the house
    * id : id of the house. You can enter into the room page by adding https://www.dabangapp.com/room/ + id             
    * loc : coordinate of each house
    * metro_count : total metro counts near each houses
    * parking: whether or not ther is parking space in the house
    * price : deposit + 10 * (rent + maintenance)
    * room_floor : floor where room is located
    * room_size : size m2 of the room
    * room_type : type of the room (one room,two room, one and half room, office-type building)
    * aircondition : airconditioner in the house        
    * laundry: laundar in the house
    * bed: bed in the house  
    * desk: desk in the house         
    * closet: closet in the house
    * tv : tv in the house
    * shoes : shoe shelf in the house
    * refrigerator : refrigerator in the house
    * gas : gas range in the house
    * microwave : microwave in the house
    * induction : induction in the house
    * doar_lock : doar lock in the house
    * biddet: biddet in the house
    * d_to_fancy : distance to closest Luxury Good Store in the Seoul
    * park : distance to closest park alongside Han River in Seoul
    * distance_gn : distance to Gangnam station
    * distance_md : distance to Myungdong station
    * d_to_mac: distance to closest Mc Donalds in Seoul
    * star_bucks : distance to closest starbucks in Seoul
    * population: population of each dong(Îèô) where the house is located in 
    * elder: total population of elderly in each dong where the house is located in 
    * elder_ratio : elder / population
    * totalbanks : total banks of each dong where the house is located in 
    * bank_ratio : totalbanks / population
