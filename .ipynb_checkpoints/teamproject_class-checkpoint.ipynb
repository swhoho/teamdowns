```python
from selenium.webdriver.support.ui import Select
import time
from pymongo import MongoClient
from selenium import webdriver
from collections import Counter

class HREF_SCRETCHER_DABANG(object):
    
    def __init__(self, dong_lst):
        dong_lst = [i.strip() for i in dong_lst] 
        self.dong_lst = dong_lst
    
   
    
    def extract_href(self):
        href_lst = []
        chromedriver = './chromedriver'
        driver = webdriver.Chrome(chromedriver)
        driver.get('https://www.dabangapp.com/search#/map?type=region&id=11140162&position=%7B%22center%22%3A%5B127.01550714548125%2C37.55690177250604%5D%2C%22zoom%22%3A14%7D&filters=%7B%22deposit-range%22%3A%5B0%2C999999%5D%2C%22price-range%22%3A%5B0%2C999999%5D%2C%22room-type%22%3A%5B0%2C1%2C2%2C3%2C4%2C5%5D%2C%22deal-type%22%3A%5B0%2C1%5D%2C%22location%22%3A%5B%5B126.97662584360137%2C37.537813508505565%5D%2C%5B127.05438844736113%2C37.57598514804419%5D%5D%7D&cluster=%7B%22name%22%3A%22%EC%8B%A0%EB%8B%B9%EB%8F%99%22%7D')
        for i in self.dong_lst:
            text_box = driver.find_element_by_xpath("//input[@class='SearchForm-input form-control']")
            text_box.clear()
            address =  i
            address = address.decode('utf8')
            text_box.send_keys(address)
            time.sleep(5)
            try:
                click_box = driver.find_element_by_xpath("//ul[@class='SearchForm-list search-items']/li/span[@class='SearchForm-item name full']")
                text = click_box.text
            except:
                print 'text_error' 
            if address in text:    
                click_box.click() 
            else:
                print 'not match'
                continue    
            time.sleep(3)
            try:
                region_box = driver.find_elements_by_class_name("Room-item")
                for i,room_info in enumerate(region_box):
                    try:
                        link = room_info.find_element_by_tag_name('a')
                        href = link.get_attribute('href')
                        href_lst.append(href)
                    except:
                        print 'no a tag with href'
                
                        
                while True:
                    try:
                        time.sleep(3)
                        next_icon = driver.find_element_by_xpath("//ul[@class='Pagination']//a[@class='Pagination-item Pagination-item--next']")
                        next_icon.click()
                        region_box = driver.find_elements_by_class_name("Room-item")
                        for i,room_info in enumerate(region_box):
                            try:
                                link = room_info.find_element_by_tag_name('a')
                                href = link.get_attribute('href')
                                href_lst.append(href)
                            except:
                                print 'no a tag with href'
                                continue
                        
                    except: 
                        print 'no next icon'
                        break

                    
            
            except:
                print 'Nothing to do'
        driver.quit()
        return href_lst
    
    def extract_final_lst(self):
        href_lst = self.extract_href()
        href_dict = Counter(href_lst)
        final_lst = href_dict.keys()
        self.final_lst = final_lst
        return final_lst
    
    def href_to_mongo(self):
        mongo = MongoClient('ec2-54-149-149-184.us-west-2.compute.amazonaws.com', 27017)
        database = mongo.Hrefs.urls
        href_data = []
        st = set()
        for i in database.find():
            st.add(i['url'])
        for i in self.final_lst:
            url = {}
            if i in st:
                continue
            url['url'] = i
            href_data.append(url)
        database.insert_many(href_data)
    def write_file(self, file_name):
        with open(file_name,'w') as txtfile:
            for i in self.final_lst:
                url = i + ' '
                txtfile.write(url)
```

* 그러던 와중에 html script를 읽자 방과 방에대한 정보가 페이지의 script안에 json데이터의 형태로 저장되어있다는 것을 알 수 있었고 json으로 되어있는 부분을 regular expression으로 긁어내어서 필요한 데이터를 얻을 수 있었다. 집과 관련된 데이터도 역시 MongoDB에 저장하여 필요할 때 업데이트를 가능하게 하였다. 

```python
from collections import Counter
import numpy as np
import pandas as pd
from pymongo import MongoClient
from operator import itemgetter
from bs4 import BeautifulSoup
import requests
import json
import re
from pymongo import MongoClient


class Href_to_data(object):
    def __init__(self):
        pass
        
    def href_to_mongodb(self):
        mongo = MongoClient('ec2-54-149-149-184.us-west-2.compute.amazonaws.com', 27017)
        hrefbase = mongo.Hrefs.urls
        house_info = mongo.dabang.house_info
        ls = []
        #이전 몽고 DB에서 존재하는 지 파악함. 
        try:
            for i in house_info.find():
                ls.append(i['id'])
        except:
            ls = []
        
        house_data = [] #각각의 data를 저장할 공간 생성.
        for href in hrefbase.find():
            web = requests.get(href['url'])
            x = re.search('dabang.web.detail\((.*}})',web.content) 
            jsonform = x.group(1)
            try:
                jsonf = json.loads(jsonform)
                housedict = {}
            except:continue
            try:
                housedict['id'] = jsonf['room']['id'] # 집 www.dabangapp.com/room/ 뒤에 아이디를 붙이면 집과 관련된 정보를 얻을 수 있음
                housedict['loc'] = (jsonf['room']['location']) # 각각의 집이 위치한 좌표
                housedict['room_size'] = jsonf['room']['room_size'] #방의 크기
                housedict['room_floor'] = jsonf['room']['room_floor'] #방의 층수
                housedict['building_floor'] = jsonf['room']['building_floor'] #건물 층수
                housedict['room_options'] = jsonf['room']['room_options'] #건물 옵션사항
            except:continue
            housedict['metro_count'] = jsonf['room']['near'][0]['total']   #주변 지하철 갯수 
            housedict['d_to_metro'] = min(jsonf['room']['near'][0]['pois'],key = itemgetter('distance'))['distance'] #가장 가까운 지하철과의 거리
            housedict['d_to_conv'] = min(jsonf['room']['near'][1]['pois'],key = itemgetter('distance'))['distance'] # 가장 가까운 편의점과의 거리
            housedict['d_to_coffe'] = min(jsonf['room']['near'][2]['pois'],key = itemgetter('distance'))['distance'] # 가장 가까운 카페와의 거리
            housedict['d_to_bank'] = min(jsonf['room']['near'][3]['pois'],key = itemgetter('distance'))['distance'] # 가장 가까운 은행과의 거리
            housedict['elevator'],housedict['parking'],housedict['animal'] = jsonf['room']['elevator'],jsonf['room']['parking'],jsonf['room']['animal']
            #가장 가까운 은행과의 거리
            try:
                if jsonf['room']['maintenance']:   #관리비
                    mc = jsonf['room']['maintenance_cost']
                else: 
                    mc = 0
                if mc is None: 
                    mc = 0
                deposit = jsonf['room']['price_info'][0][0] #보증금 
                if deposit is None: 
                    deposit = 0
                rent = jsonf['room']['price_info'][0][1] #월세
                if rent is None :
                    rent = 0
                price = deposit + (rent + mc) * 100
            except:continue
            housedict['price'] = price #최종가격
            house_data.append(housedict)
        house_info.insert_many(house_data)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder,LabelBinarizer
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline      import FeatureUnion
from sklearn.preprocessing import RobustScaler
from sklearn  import metrics 
#from sklearn.preprocessing import ColumnExtractor
class Price_extractor(object):
    def __init__(self):
        pass
   
    
    def transform(self,df):
        for i in df: 
            if i =='price':
                return df.price.reshape(-1,1)
    
    def fit(self,df,y = None):
        return self
class str_extractor(object):
    def __init__(self):
        pass
    def transform(self,df):
        lst = []
        for i in df: 
            if type(df[i][0]) ==str or type(df[i][0]) == unicode :
                lst.append(i)
        
        return df[lst]
    
    def fit(self,df,y = None):
        return self
class float_extractor(object):
    
    def __init__(self):
        pass
    
    def transform(self,df):
        lst = []
        for i in df: 
            try:
                if type(df[i][0]) ==float or np.mean(df[i]) > 1:
                    lst.append(i)
            except: 
                continue

        return df[lst]
    
    def fit(self,df,y = None):
        return self
class bool_extractor(object):
    
    def __init__(self):
        pass
    
    def transform(self,df):
        lst = []
        for i in df: 
            try:
                if type(df[i][0]) =='numpy.bool_':
                    lst.append(i)
            except: 
                continue

        return df[lst]
    
    def fit(self,df,y = None):
        return self
class DenseTransformer(object):

    def transform(self, X, y=None, **fit_params):
        return X.todense()

    def fit_transform(self, X, y=None, **fit_params):
        self.fit(X, y, **fit_params)
        return self.transform(X)

    def fit(self, X, y=None, **fit_params):
        return self
class MultiColumnLabelEncoder(object):
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
lb = LabelBinarizer()    
pf = PolynomialFeatures(2,interaction_only = True)
le = LabelEncoder()
ohe = OneHotEncoder()
model = LogisticRegression()
sc  = StandardScaler()
dt = DenseTransformer()
from sklearn.preprocessing import Imputer
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LinearRegression
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
pl = Pipeline([('FeatureUnion', FeatureUnion(
                                            [('characters' , Pipeline([('change_str',str_extractor()),
                                                                    ('Label_encoding',MultiColumnLabelEncoder()),
                                                                     ('One_Hot_Encoding',ohe),
                                                                      ('Dense_transform',dt)])),
                                         
                                            ('real_number',Pipeline([('float_extractor',float_extractor()),
                                                                     ('imputer',imp),
                                                                     ('poly_nomial', pf),
                                                                     ('standar_scaler', Normalizer())]))
                                            ])
               )])
features = df[[ u'animal', u'building_floor',
            u'd_to_bank',     u'd_to_coffe',      u'd_to_conv',
           u'd_to_metro',       u'elevator',    
            u'metro_count',        u'parking',
             u'room_floor',
            u'room_size',      u'room_type',    'd_to_fancy',
         u'aircondition',        u'laundry',           # u'bed',
                 u'desk',        # u'closet',             u'tv',
                u'shoes',   u'refrigerator',            u'gas',
                  u'microwave',      u'doar_lock',
               u'biddet',           u'park',    u'distance_gn',
          u'distance_md',       u'd_to_mac',     u'star_bucks']]
transformed_features =  pl.fit_transform(features)
model = LogisticRegression()
#cross_val_score(model, transformed_features, df.induction, "r2")
import matplotlib.pylab as plt
model = LogisticRegression().fit(X = df.room_floor.reshape(-1,1), y = df.induction.reshape(-1,1))
xx = np.linspace(-3, 200, 6577)
sigm = 1.0/(1 + np.exp(-model.coef_[0][0]*xx - model.intercept_[0]))
plt.plot(xx, sigm)
plt.hold(True)
#plt.scatter(df.room_size.reshape,df.induction, marker='o', c='y', s=10)
plt.show()
print model.coef_[0][0]

```
