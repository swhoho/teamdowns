* Selenium에서 데이터의 URL을 수집 
 * elements와 element차이점 구분 주의 
 * 적절한 time.sleep기능을 이용해 너무 빠르거나 느리게 지나치지 않도록 설정 

```python
from selenium.webdriver.support.ui import Select
import time
from pymongo import MongoClient
from selenium import webdriver
from collections import Counter
class HREF_SCRETCHER_DABANG(object):
    
    def __init__(self, dong_lst):
        dong_lst = [i.strip() for i in dong_lst] 
        self.dong_lst = dong_lst
    
   
    
    def extract_href(self):
        href_lst = []
        chromedriver = './chromedriver'
        driver = webdriver.Chrome(chromedriver)
        driver.get('https://www.dabangapp.com/search#/map?type=region&id=11140162&position=%7B%22center%22%3A%5B127.01550714548125%2C37.55690177250604%5D%2C%22zoom%22%3A14%7D&filters=%7B%22deposit-range%22%3A%5B0%2C999999%5D%2C%22price-range%22%3A%5B0%2C999999%5D%2C%22room-type%22%3A%5B0%2C1%2C2%2C3%2C4%2C5%5D%2C%22deal-type%22%3A%5B0%2C1%5D%2C%22location%22%3A%5B%5B126.97662584360137%2C37.537813508505565%5D%2C%5B127.05438844736113%2C37.57598514804419%5D%5D%7D&cluster=%7B%22name%22%3A%22%EC%8B%A0%EB%8B%B9%EB%8F%99%22%7D')
        for i in self.dong_lst:
            text_box = driver.find_element_by_xpath("//input[@class='SearchForm-input form-control']")
            text_box.clear()
            address =  i
            address = address.decode('utf8')
            text_box.send_keys(address)
            time.sleep(5)
            try:
                click_box = driver.find_element_by_xpath("//ul[@class='SearchForm-list search-items']/li/span[@class='SearchForm-item name full']")
                text = click_box.text
            except:
                print 'text_error' 
            if address in text:    
                click_box.click() 
            else:
                print 'not match'
                continue    
            time.sleep(3)
            try:
                region_box = driver.find_elements_by_class_name("Room-item")
                for i,room_info in enumerate(region_box):
                    try:
                        link = room_info.find_element_by_tag_name('a')
                        href = link.get_attribute('href')
                        href_lst.append(href)
                    except:
                        print 'no a tag with href'
                
                        
                while True:
                    try:
                        time.sleep(3)
                        next_icon = driver.find_element_by_xpath("//ul[@class='Pagination']//a[@class='Pagination-item Pagination-item--next']")
                        next_icon.click()
                        region_box = driver.find_elements_by_class_name("Room-item")
                        for i,room_info in enumerate(region_box):
                            try:
                                link = room_info.find_element_by_tag_name('a')
                                href = link.get_attribute('href')
                                href_lst.append(href)
                            except:
                                print 'no a tag with href'
                                continue
                        
                    except: 
                        print 'no next icon'
                        break

                    
            
            except:
                print 'Nothing to do'
        driver.quit()
        return href_lst
    
    def extract_final_lst(self):
        href_lst = self.extract_href()
        href_dict = Counter(href_lst)
        final_lst = href_dict.keys()
        self.final_lst = final_lst
        return final_lst
    
    def href_to_mongo(self):
        mongo = MongoClient('ec2-54-149-149-184.us-west-2.compute.amazonaws.com', 27017)
        database = mongo.Hrefs.urls
        href_data = []
        st = set()
        for i in database.find():
            st.add(i['url'])
        for i in self.final_lst:
            url = {}
            if i in st:
                continue
            url['url'] = i
            href_data.append(url)
        database.insert_many(href_data)
    def write_file(self, file_name):
        with open(file_name,'w') as txtfile:
            for i in self.final_lst:
                url = i + ' '
                txtfile.write(url)
```

```python
from bs4 import BeautifulSoup
import requests
import json
import re
import unicodecsv as csv
from pymongo import MongoClient

class Href_to_data(object):
    def __init__(self,href_lst):
        self.href_lst = href_lst
         
    def checkout_duplicated(self,fname):
        df = pd.read_csv(fname)
        url = 'https://www.dabangapp.com/room/'
        copy = url + df['id'] 
        copy = set(copy)
        dong_lst = set(self.dong_lst)
        return dong_lst - copy
    
    def href_to_mongodb(self):
        mongo = MongoClient('ec2-54-149-149-184.us-west-2.compute.amazonaws.com', 27017)
        house_info = mongo.dabang.house_info
        house_data = []
        for href in self.href_lst:
            web = requests.get(href)
            x = re.search('dabang.web.detail\((.*}})',web.content) 
            jsonform = x.group(1)
            jsonf = json.loads(jsonform)
            housedict = {}
            housedict['loc'] = (jsonf['room']['location'])
            housedict['room_size'] = jsonf['room']['room_size']
            housedict['room_floor'] = jsonf['room']['room_floor']
            housedict['building_floor'] = jsonf['room']['building_floor']
            housedict['elevator'] = jsonf['room']['elevator']
            housedict['parking'] = jsonf['room']['parking']
            housedict['animal'] = jsonf['room']['animal']
            housedict['room_options'] = jsonf['room']['room_options']
            housedict['id'] = jsonf['room']['id']
            if jsonf['room']['maintenance']:
                mc = jsonf['room']['maintenance_cost']
            else: 
                mc = 0
            if mc is None: 
                mc = 0
            deposit = jsonf['room']['price_info'][0][0] 
            if deposit is None: 
                deposit = 0
            rent = jsonf['room']['price_info'][0][1] 
            if rent is None :
                rent = 0
            price = deposit + (rent + mc) * 100
            housedict['price'] = price
            house_data.append(housedict)
        house_info.insert_many(house_data)
    
    def href_to_csv(self,fname):
        with open(fname, 'w') as fw:
            writer = csv.writer(fw)
            writer.writerow(['id','loc','room_type','size', 'room_floor','building_floor','parking','elevator'
                         ,'heat_type','animal','aircondition','laundry','bed','desk','closet','tv'
                         ,'shoes','refrigerator','gas','induction','microwave','doar_lock','bidet'
                        ,'price'])
            for href in self.href_lst:
                web = requests.get(href)
                x = re.search("dabang.web.detail\((.*}})",web.content)
                jsonform = x.group(1)
                jsonf = json.loads(jsonform)
                loc = jsonf['room']['location']
                room_type =  jsonf['room']['room_type_str']
                size = jsonf['room']['room_size']
                room_floor = jsonf['room']['room_floor']
                building_floor = jsonf['room']['building_floor']
                elevator = jsonf['room']['elevator']
                parking = jsonf['room']['parking']
                heat_type = jsonf['room']['heating']
                animal = jsonf['room']['animal']
                options = jsonf['room']['room_options']
                room_id = jsonf['room']['id']
                aircondition = 0 in options
                laundry = 1 in options
                bed = 2 in options
                desk = 3 in options
                closet = 4 in options
                tv = 5 in options
                shoes = 6 in options
                refrigerator = 7 in options
                gas = 8 in options
                induction = 9 in options
                microwave = 10 in options
                doar_lock = 11 in options
                bidet = 12 in options
                if jsonf['room']['maintenance']:
                    mc = jsonf['room']['maintenance_cost']
                else: 
                    mc = 0
                if mc is None: 
                    mc = 0
                deposit = jsonf['room']['price_info'][0][0] 
                if deposit is None: 
                    deposit = 0
                rent = jsonf['room']['price_info'][0][1] 
                if rent is None :
                    rent = 0
                price = deposit + (rent + mc) * 100
                data_row = [room_id, loc, room_type, size, room_floor, building_floor, parking,
                                 elevator, heat_type, animal, aircondition, laundry, bed, desk,
                                 closet,tv, shoes, refrigerator, gas,induction, microwave,
                                 doar_lock,bidet,price]
                self.fname = fname
                writer.writerow(data_row)
                
    def csv_to_df(self):
        df = pd.read_csv(self.fname)
        return df
```

```python
import pandas as pd
import numpy as np
import requests
import json
class Transform_csv(object):
    
  
    def __init__(self,fname):
        df = pd.read_csv(fname)
        self.df = df
        location=  [re.findall('[0-9.]+',i) for i in  self.df['loc']]
        X= [float(i[0]) for i in location]
        Y = [float(i[1]) for i in location]
        only_loc = np.hstack([np.asarray(X).reshape(-1,1),np.asarray(Y).reshape(-1,1)])
        self.only_loc = only_loc
    park = [[126.8141241,37.5876169], [126.8979808,37.5397924], [126.9939698,37.5102684], [127.0165227,37.5272945], [127.081523,37.5179568]
                 ,[127.1194287,37.5499881],[126.8721985,37.5649224],  [126.8986955,37.555875], [126.9736383,37.5159618], [127.0691564,37.5291008]]

    
    def location_point(self,lst):
        loc_list = []
        for depart in lst: 
            depart = depart.decode('utf-8').encode('utf-8')
            naverurl =  'https://openapi.naver.com/v1/map/geocode?encoding=utf-8&coord=latlng&output=json&query={}'.format(depart)
            headers = {}
            headers['X-Naver-Client-Id'] =  'fMc7n_JcnZoItjRY0HHp'
            headers['X-Naver-Client-Secret'] = 'agQGiCxnNF'
            res = requests.get(naverurl, headers = headers)
            jsonf = json.loads(res.content)
            loc_list.append([float(jsonf['result']['items'][0]['point']['x']),float(jsonf['result']['items'][0]['point']['y']) ])
        return loc_list
        
    def d_to_places(self,loc_lst):

        def dist_places(x):
            z = []
            for i in range(len(loc_lst)):
                k = ((114*(x[0] - loc_lst[i][0]))**2 + ((92*(x[1] - loc_lst[i][1]))**2))**(1.0/2.0)
                z.append(k)
            return min(z)
        return np.apply_along_axis(dist_places,1,self.only_loc)

    def d_to_park(self):

        def dist_park(x):
            z = []
            for i in range(len(park)):
                k = ((114*(x[0] - park[i][0]))**2 + ((92*(x[1] - park[i][1]))**2))**(1.0/2.0)
                z.append(k)
            return min(z)
        return np.apply_along_axis(dist_park,1,self.only_loc)
        
  

    def d_to_myoungdong(self):
        def dist_myongdong(x):
            k = ((114*(x[0] - 126.9653337))**2 + ((92*(x[1] - 37.5634056))**2))**(1.0/2.0)
            return k
        return np.apply_along_axis(dist_myongdong,1,self.only_loc)
    
    
    def d_to_gangnam(self):    
        def dist_gangnam(x):
            k = ((114*(x[0] - 127.0083444))**2 +(92*(x[1] - 37.5002943))**2)**(1.0/2.0) 
            return k
        return np.apply_along_axis(dist_gangnam,1,self.only_loc)
```

PipeLine을 통해 DataFrame의 전처리기능을 한번에 처리할 수 있는 툴을 만들어봄

```python
from sklearn import cross_validation 
import pandas as pd
import numpy as np
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline      import FeatureUnion
from sklearn.preprocessing import RobustScaler
from sklearn  import metrics 

#from sklearn.preprocessing import ColumnExtractor
class Price_extractor(object):
    def __init__(self):
        pass
   
    
    def transform(self,df):
        for i in df: 
            if i =='price':
                return df.price.reshape(-1,1)
    
    def fit(self,df,y = None):
        return self
class str_extractor(object):
    def __init__(self):
        pass
    def transform(self,df):
        lst = []
        for i in df: 
            if type(df[i][0]) ==str:
                lst.append(i)
        return df[lst]
    
    def fit(self,df,y = None):
        return self
class float_extractor(object):
    
    def __init__(self):
        pass
    
    def transform(self,df):
        lst = []
        for i in df: 
            try:
                if type(df[i][0]) ==float or np.mean(df[i]) > 1:
                    lst.append(i)
            except: 
                continue

        return df[lst]
    
    def fit(self,df,y = None):
        return self
class DenseTransformer(object):

    def transform(self, X, y=None, **fit_params):
        return X.todense()

    def fit_transform(self, X, y=None, **fit_params):
        self.fit(X, y, **fit_params)
        return self.transform(X)

    def fit(self, X, y=None, **fit_params):
        return self
class MultiColumnLabelEncoder(object):
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
pf = PolynomialFeatures(3)
le = LabelEncoder()
ohe = OneHotEncoder()
model = LinearRegression()
sc  = StandardScaler()
dt = DenseTransformer()
pl = Pipeline([('FeatureUnion', FeatureUnion(
                                            [('characters' , Pipeline([('change_str',str_extractor()),
                                                                    ('Label_encoding',MultiColumnLabelEncoder()),
                                                                     ('One_Hot_Encoding',ohe),
                                                                      ('Dense_transform',dt)]),),
                                            ('real_number',Pipeline([('float_extractor',float_extractor()),
                                                                     ('poly_nomial', pf),
                                                                     ('standar_scaler', Normalizer())]))
                                            ])
               )])
df = pd.read_csv('updated_fancy.csv')
sr = str_extractor()
features =  pl.fit_transform(df.ix[:,3:-1])
cv = cross_validation.KFold(len(df.price),n_folds= 5)
log_p = np.log(df.price)
for train_idx, test_idx in cv:
    features_train = features[train_idx]
    features_test = features[test_idx]
    target_train = log_p[train_idx]
    target_test= log_p[test_idx]
    model.fit(features_train, target_train)
    target_pred = model.predict(features_test)
    print metrics.r2_score(target_test,target_pred)
pl2 =Pipeline([('float_extractor',float_extractor()),
                                         ('standar_scaler', sc)])
```
